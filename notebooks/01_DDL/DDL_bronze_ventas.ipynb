```python
%sql

-- Cracion de esquema bronze

USE iniciacion_deportiva.bronze;

USE SCHEMA bronze;

CREATE OR REPLACE TABLE ventas_raw (
    -- Todo como STRING para máxima flexibilidad
    id_venta STRING,
    fecha STRING, 
    id_producto STRING,
    titulo_producto STRING,
    cantidad STRING,       
    categoria_id STRING,
    precio_unitario STRING,
    comision STRING,        
    id_cliente STRING,
    cliente_nickname STRING,
    estado STRING,
    _source STRING COMMENT 'origen del dato',
    ingested_at TIMESTAMP COMMENT 'tiempo de ingesta'
) USING DELTA;
```
