```python
%sql
-- ============================================================
-- dim_producto — SCD Type 2 sobre titulo_producto
-- Cada cambio de título genera una nueva fila.
-- El precio NO se trackea acá — vive en fact_ventas.
-- ============================================================
CREATE TABLE IF NOT EXISTS iniciacion_deportiva.gold.dim_producto (
    producto_key        BIGINT          GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    id_producto         STRING          NOT NULL,   -- natural key Meli
    titulo_producto     STRING          NOT NULL,   -- SCD Type 2 — dispara nueva fila si cambia
    categoria_id        STRING,
    nivel_1             STRING,                     -- desnormalizado desde silver.categorias
    nivel_2             STRING,
    nombre_categoria    STRING,                     -- nombre hoja de la categoría
    fecha_inicio        DATE            NOT NULL,   -- desde cuándo aplica este título
    fecha_fin           DATE            NOT NULL,   -- 9999-12-31 = fila activa
    es_actual           BOOLEAN         NOT NULL,   -- TRUE = fila activa
    _source STRING DEFAULT 'iniciacion_deportiva.silver.ventas', -- Origen
    _created_at         TIMESTAMP       DEFAULT CURRENT_TIMESTAMP(),
    _updated_at         TIMESTAMP       DEFAULT CURRENT_TIMESTAMP()
) USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Dimensión de productos - SCD Type 2 sobre titulo_producto';
```
