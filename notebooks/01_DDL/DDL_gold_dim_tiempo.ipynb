```python
%sql
-- ============================================================
-- dim_tiempo — tabla de fechas estática
-- Se genera una fila por día desde el min al max de fact_ventas.
-- El ETL la recrea completa cada vez (CREATE OR REPLACE).
-- ============================================================
CREATE TABLE IF NOT EXISTS iniciacion_deportiva.gold.dim_tiempo (
    tiempo_key          INT             NOT NULL,   -- YYYYMMDD como entero, ej: 20250315
    fecha               DATE            NOT NULL,
    dia                 INT             NOT NULL,   -- 1-31
    mes                 INT             NOT NULL,   -- 1-12
    nombre_mes          STRING          NOT NULL,   -- 'Enero', 'Febrero', etc.
    trimestre           INT             NOT NULL,   -- 1-4
    semestre            INT             NOT NULL,   -- 1-2
    anio                INT             NOT NULL,
    _created_at         TIMESTAMP       DEFAULT CURRENT_TIMESTAMP()
) USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Dimensión de tiempo — 1 fila por día, rango dinámico desde silver.ventas';
```
