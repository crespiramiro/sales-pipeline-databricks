```python
%sql
-- ============================================================
-- fact_ventas — granularidad: 1 fila = 1 línea de venta
-- FKs a dim_producto (is_current) y dim_tiempo.
-- total_linea calculado acá, no en Silver.
-- estado incluido para análisis de cancelaciones futuro.
-- ============================================================
CREATE TABLE IF NOT EXISTS iniciacion_deportiva.gold.fact_ventas (
    venta_key           BIGINT          GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    producto_key        BIGINT          NOT NULL,   -- FK → dim_producto
    tiempo_key          INT             NOT NULL,   -- FK → dim_tiempo
    pedido_key          STRING ,-- clave degenerada (id_cliente + minuto)
    id_venta            STRING          NOT NULL,   -- natural key, trazabilidad
    id_cliente          STRING,                     -- sin dim — mejora futura
    precio_unitario     DOUBLE          NOT NULL,
    cantidad            INT             NOT NULL,
    comision            DOUBLE,
    total_venta         DOUBLE          NOT NULL,   -- precio_unitario * cantidad
    estado              STRING          NOT NULL,   -- paid | cancelled — para análisis futuro
    _source       STRING          DEFAULT 'iniciacion_deportiva.silver.ventas',
    _processing_timestamp TIMESTAMP     DEFAULT CURRENT_TIMESTAMP()
) USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Fact de ventas — granularidad 1 fila por línea de venta';
```
