```python
%sql
CREATE TABLE IF NOT EXISTS iniciacion_deportiva.silver.categorias (
    categoria_id      STRING        NOT NULL,  -- PK
    nombre            STRING,                  -- nombre de la categoría hoja
    nivel_1           STRING,                  -- "Deportes y Fitness"
    nivel_2           STRING,                  -- "Natación"
    nivel_3           STRING,                  -- "Gorras de Natación"
    nivel_4           STRING,                  -- puede ser NULL
    nivel_5           STRING,                  -- puede ser NULL
    path_completo     STRING,                  -- "Deportes > Natación > ..."
     -- Metadata de procesamiento
    _source STRING DEFAULT 'bootcamp.bronze.categorias_raw' COMMENT 'Tabla origen',
    _processing_timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP() COMMENT 'Timestamp de procesamiento'       
) USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Categorias - Capa Silver (datos limpios y validados)';
```
