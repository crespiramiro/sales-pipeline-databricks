```python
%sql
CREATE TABLE IF NOT EXISTS iniciacion_deportiva.silver.ventas (
    id_venta          STRING        NOT NULL,  -- PK — deduplicado desde Bronze
    fecha             TIMESTAMP     NOT NULL,  
    id_producto       STRING        NOT NULL,
    titulo_producto   STRING,                  -- título más reciente del listing
    cantidad          INT           NOT NULL,
    categoria_id      STRING,
    precio_unitario   DOUBLE        NOT NULL,
    comision          DOUBLE,
    id_cliente        STRING,
    cliente_nickname  STRING,
    estado            STRING        NOT NULL,  -- paid | partially_refunded
    -- Metadata de procesamiento
    _source STRING DEFAULT 'bootcamp.bronze.ventas_raw' COMMENT 'Tabla origen',
    _processing_timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP() COMMENT 'Timestamp de procesamiento'          
) USING DELTA
 TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
 COMMENT 'Ventas - Capa Silver (datos limpios y validados)';
```
