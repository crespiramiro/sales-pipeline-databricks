 ## 🥉 Maintenance — Bronze Layer
Compacta archivos fragmentados y limpia versiones viejas.
Bronze es la capa más fragmentada: 1 archivo Parquet por corrida de GitHub Actions (~1/hora).

```python
%sql

-- Estado antes

DESCRIBE DETAIL iniciacion_deportiva.bronze.ventas_raw;
```

```python
%sql
-- No usamos ZORDER en Bronze porque no se consulta directamente en el dashboard.
-- El objetivo es solo reducir la cantidad de archivos para que Auto Loader
-- y los ETL Silver no tengan que escanear miles de archivos chicos.
-- OPTIMIZE

OPTIMIZE iniciacion_deportiva.bronze.ventas_raw;
```

```python
%sql
-- OPTIMIZE categorias_raw
OPTIMIZE iniciacion_deportiva.bronze.categorias_raw;
```

```python
%sql
-- VACUUM
-- Retención: 168 horas (7 días) — mínimo recomendado por Databricks.
-- No bajar de este valor: rompe time travel y puede causar errores en lecturas concurrentes con Auto Loader.

VACUUM iniciacion_deportiva.bronze.ventas_raw RETAIN 168 HOURS;
```

```python
%sql
--VACUUM categorias_raw
VACUUM iniciacion_deportiva.bronze.categorias_raw RETAIN 168 HOURS;
```

```python
%sql
--Estado DESPUÉS del mantenimiento
DESCRIBE DETAIL iniciacion_deportiva.bronze.ventas_raw;
```
